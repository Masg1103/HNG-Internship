```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Customer Analysis Visualization

```python
# Load your CSV files
top_5_customers_df = pd.read_csv("Customer Analysis - Top 5 customers who have spent the most on vehicle repairs and parts.csv")
average_spending_df = pd.read_csv("Customer Analysis - Average Spending.csv")
customer_visit_frequency_df = pd.read_csv('Customer Analysis - frequency of customer visits and identify any patterns.csv')
```

```python
top_5_customers_df.head()
```

```python
average_spending_df.head()
```

```python
customer_visit_frequency_df.head()
```

```python
# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(24, 6))

# Top 5 Customers by Spending (Pie Chart)
axs[0].pie(top_5_customers_df['TotalSpending'], labels=top_5_customers_df['CustomerName'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green', 'red', 'purple'])
axs[0].set_title('Top 5 Customers by Total Spending')

# Average Spending per Customer
axs[1].bar(['Average Spending'], average_spending_df['AverageSpendingPerCustomer'], color='orange')
axs[1].set_title('Average Spending per Customer')
axs[1].set_ylabel('Average Spending ($)')

# Customer Visit Frequency
axs[2].bar(customer_visit_frequency_df['CustomerID'], customer_visit_frequency_df['VisitFrequency'], color='green')
axs[2].set_title('Customer Visit Frequency')
axs[2].set_xlabel('Customer ID')
axs[2].set_ylabel('Visit Frequency')

plt.tight_layout()
plt.savefig('customer_analysis_combined.png')
plt.show()
```

### Vehicle Analysis Visualization

```python
# Load your CSV files
average_mileage_df = pd.read_csv('Vehicle Analysis - Average mileage.csv')
most_common_vehicles_df = pd.read_csv('Vehicle Analysis - most common vehicle makes and models brought in for service.csv')
vehicle_ages_distribution_df = pd.read_csv('Vehicle Analysis - The distribution of vehicle ages.csv')
service_requirements_df = pd.read_csv('Vehicle Analysis - the distribution of vehicle ages and identify trend in service requirements based on vehicle age.csv')
```

```python
average_mileage_df
```

```python
most_common_vehicles_df
```

```python
vehicle_ages_distribution_df
```

```python
service_requirements_df
```

```python
# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(28, 14))

# Average Mileage of Vehicles Serviced
axs[0, 0].bar(['Average Mileage'], average_mileage_df['AverageMileage'], color='blue')
axs[0, 0].set_title('Average Mileage of Vehicles Serviced')
axs[0, 0].set_ylabel('Average Mileage')

# Most Common Vehicle Makes and Models
axs[0, 1].bar(most_common_vehicles_df['Make'] + ' ' + most_common_vehicles_df['Model'], most_common_vehicles_df['VehicleCount'], color='green')
axs[0, 1].set_title('Most Common Vehicle Makes and Models Brought in for Service')
axs[0, 1].set_xlabel('Vehicle Make and Model')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].set_xticklabels(most_common_vehicles_df['Make'] + ' ' + most_common_vehicles_df['Model'], rotation=45, ha='right')

# Distribution of Vehicle Ages (Pie Chart)
axs[1, 0].pie(vehicle_ages_distribution_df['VehicleCount'], labels=vehicle_ages_distribution_df['AgeGroup'], autopct='%1.1f%%', startangle=90, colors=['red', 'orange', 'yellow', 'green'])
axs[1, 0].set_title('Distribution of Vehicle Ages')

# Service Requirements Based on Vehicle Age
job_descriptions = []
for key, grp in service_requirements_df.groupby(['AgeGroup']):
    axs[1, 1].bar(grp['Description'], grp['JobCount'], label=key)
    job_descriptions.extend(grp['Description'])
axs[1, 1].set_title('Service Requirements Based on Vehicle Age')
axs[1, 1].set_xlabel('Job Description')
axs[1, 1].set_ylabel('Job Count')
axs[1, 1].legend()
axs[1, 1].set_xticklabels(job_descriptions, rotation=45, ha='right')


plt.tight_layout(pad=3.0)
plt.savefig('vehicle_analysis_combined_with_pie_and_rotated_labels.png')
plt.show()
```

### Job Performed Analysis Visualization

```python
# Load your CSV files
common_jobs_df = pd.read_csv('Job Performed Analysis -most common types of jobs performed and their frequency.csv')
total_revenue_jobs_df = pd.read_csv('Job Performed Analysis - the total revenue generated from each type of job.csv')
total_revenue_all_jobs_df = pd.read_csv('Job Performed Analysis - the total revenue generated -all jobs combined.csv')
highest_lowest_cost_jobs_df = pd.read_csv('Job Performed Analysis - jobs with the highest and lowest average costs.csv')

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(24, 12))

# Most Common Types of Jobs Performed
axs[0, 0].bar(common_jobs_df['Description'], common_jobs_df['Frequency'], color='skyblue')
axs[0, 0].set_title('Most Common Types of Jobs Performed')
axs[0, 0].set_xlabel('Job Description')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].set_xticklabels(common_jobs_df['Description'], rotation=45, ha='right')

# Total Revenue Generated from Each Type of Job (Bar and Line Plot)
axs[0, 1].bar(total_revenue_jobs_df['Description'], total_revenue_jobs_df['TotalRevenue'], color='orange', alpha=0.6, label='Revenue')
ax2 = axs[0, 1].twinx()
ax2.plot(total_revenue_jobs_df['Description'], total_revenue_jobs_df['TotalRevenue'], color='blue', marker='o', linestyle='-', label='Revenue Line')
axs[0, 1].set_title('Total Revenue Generated from Each Type of Job')
axs[0, 1].set_xlabel('Job Description')
axs[0, 1].set_ylabel('Total Revenue ($)')
axs[0, 1].set_xticklabels(total_revenue_jobs_df['Description'], rotation=45, ha='right')
ax2.set_ylabel('Total Revenue ($)')
axs[0, 1].legend(loc='upper left')
ax2.legend(loc='upper right')

# Total Revenue from All Jobs Combined
axs[1, 0].bar(['Total Revenue'], total_revenue_all_jobs_df['TotalRevenue'], color='green')
axs[1, 0].set_title('Total Revenue from All Jobs Combined')
axs[1, 0].set_ylabel('Total Revenue ($)')

# Jobs with the Highest and Lowest Average Costs
axs[1, 1].bar(highest_lowest_cost_jobs_df['Description'], highest_lowest_cost_jobs_df['AverageCost'], color='red')
axs[1, 1].set_title('Jobs with the Highest and Lowest Average Costs')
axs[1, 1].set_xlabel('Job Description')
axs[1, 1].set_ylabel('Average Cost ($)')
axs[1, 1].set_xticklabels(highest_lowest_cost_jobs_df['Description'], rotation=45, ha='right')


plt.tight_layout()
plt.savefig('job_performance_analysis_combined.png')
plt.show()
```

### Part Usage Analysis Visualization

```python
# Load your CSV files
average_part_cost_df = pd.read_csv('Part Usage Analysis - the average cost of parts used in repairs.csv')
top_5_parts_df = pd.read_csv('Part Usage Analysis - top 5 most frequently used parts and their total usage.csv')
total_parts_revenue_df = pd.read_csv('Part Usage Analysis - Total Revenue from part sales.csv')

# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(24, 6))

# Top 5 Most Frequently Used Parts
axs[0].bar(top_5_parts_df['PartName'], top_5_parts_df['TotalQuantity'], color='skyblue')
axs[0].set_title('Top 5 Most Frequently Used Parts')
axs[0].set_xlabel('Part Name')
axs[0].set_ylabel('Total Quantity Used')
axs[0].set_xticklabels(top_5_parts_df['PartName'], rotation=45, ha='right')

# Average Cost of Parts Used in Repairs
axs[1].bar(['Average Part Cost'], average_part_cost_df['AveragePartCost'], color='orange')
axs[1].set_title('Average Cost of Parts Used in Repairs')
axs[1].set_ylabel('Average Cost ($)')

# Total Revenue from Parts Sales
axs[2].bar(['Total Revenue'], total_parts_revenue_df['TotalPartsRevenue'], color='green')
axs[2].set_title('Total Revenue from Parts Sales')
axs[2].set_ylabel('Total Revenue ($)')

plt.tight_layout()
plt.savefig('parts_usage_analysis_combined.png')
plt.show()
```

### Financial Analysis Visualization

```python
# Load your CSV files
impact_of_sales_tax_df = pd.read_csv('Financial Analysis- Impact of sales tax on the total revenue.csv')
overall_profitability_df = pd.read_csv('Financial Analysis- Overall profitability.csv')
revenue_labour_parts_df = pd.read_csv('Financial Analysis- Total Revenue for labour and parts for each month.csv')

# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(24, 6))

# Total Revenue Generated from Labor and Parts for Each Month
axs[0].plot(revenue_labour_parts_df['Month'], revenue_labour_parts_df['TotalLaborRevenue'], label='Labor Revenue', marker='o', color='blue')
axs[0].plot(revenue_labour_parts_df['Month'], revenue_labour_parts_df['TotalPartsRevenue'], label='Parts Revenue', marker='o', color='green')
axs[0].set_title('Total Revenue Generated from Labor and Parts for Each Month')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Revenue ($)')
axs[0].legend()

# Overall Profitability
axs[1].bar(['Overall Profit'], overall_profitability_df['OverallProfit'], color='orange')
axs[1].set_title('Overall Profitability')
axs[1].set_ylabel('Profit ($)')

# Impact of Sales Tax on Total Revenue
axs[2].bar(['Total Sales Tax', 'Total Revenue'], [impact_of_sales_tax_df['TotalSalesTax'][0], impact_of_sales_tax_df['TotalRevenue'][0]], color=['red', 'purple'])
axs[2].set_title('Impact of Sales Tax on Total Revenue')
axs[2].set_ylabel('Amount ($)')

plt.tight_layout()
plt.savefig('financial_analysis_combined.png')
plt.show()
```
